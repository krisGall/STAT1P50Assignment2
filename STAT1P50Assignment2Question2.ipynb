```python
from random import sample

for i in range(0,100):
    a = sample(['1','2','3'], 1)
    b = sample(['1','2','3'], 1)
    
    file_a = open(file='file1.txt', mode='a')
    file_b = open(file='file2.txt', mode='a')
    
    file_a.write('\n'+a[0])
    file_b.write('\n'+b[0])
    
    file_a.close()
    file_b.close()
    
    # reading files
    file_a = open(file='file1.txt', mode='r')
    file_b = open(file='file2.txt', mode='r')
    
    # storing lines in variables
    file_a_lines = file_a.readlines()
    file_b_lines = file_b.readlines()
    
    # checking if lines are equal
    for line in range(0,len(file_a_lines)):
        if file_a_lines[line] == file_b_lines[line]:
            print("Success!")
    
```
