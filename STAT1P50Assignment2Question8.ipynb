```python
import pandas
import matplotlib.pyplot
import seaborn

# importing csv
data = pandas.read_csv('student.csv')

# setting data to studytime and grade data
data = data[['studytime', 'grade']]

# splitting data into studytime > 3 and studytime >= 3
studytimeLessThanThree = data[data['studytime'] < 3]
studytimeGrtrThanEqualThree = data[data['studytime'] >= 3]

# recreating dataframe with split data
data = pandas.DataFrame({'Study Time Less than 3': studytimeLessThanThree['grade'],
                         'Study Time Greater than or Equal to 3': studytimeGrtrThanEqualThree['grade']})

# showing boxplots
seaborn.boxplot(data=data)
```
