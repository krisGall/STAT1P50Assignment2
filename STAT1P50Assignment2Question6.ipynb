```python
import numpy
import pandas
import matplotlib.pyplot
from sklearn.preprocessing import StandardScaler

# Creating data frame
data = pandas.read_csv('crime.csv', 
                       delimiter = ',')

# Reshaping data
pctPopUnderPov = numpy.array(data['PctPopUnderPov']).reshape(-1,1)
violentCrimesPerPop = numpy.array(data['ViolentCrimesPerPop']).reshape(-1,1)

# Outputting scatter plot
print(matplotlib.pyplot.scatter(pctPopUnderPov, 
                                violentCrimesPerPop))
```

```python
# Creating scaler object
scaler = StandardScaler()

# Scaling data
pctPopUnderPovScaled = scaler.fit_transform(pctPopUnderPov)
violentCrimesPerPopScaled = scaler.fit_transform(violentCrimesPerPop)

# Outputting scatter plot
print(matplotlib.pyplot.scatter(pctPopUnderPovScaled,
                                violentCrimesPerPopScaled))
```
